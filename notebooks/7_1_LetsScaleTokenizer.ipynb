```python
from datasets import load_dataset

# Load a specific language subset (recommended — full dataset is huge)
ds = load_dataset("HuggingFaceFW/fineweb-2", name="ita_Latn", split="train", streaming=True)
```

## Load the sequence

We'll load a sample from the text sequence instead of the entire dataset to prevent excessive RAM usage. If the RAM is full, the BPE algorithm won't function properly due to a lack of available memory.

Adjust the `number_of_characters_to_read` value to find the optimal setting for your system.

```python
TARGET_CHARS = 20_000_000
MIN_CHARS    = 100
MAX_CHARS    = 100_000

parts = []
total_chars = 0

for sample in ds:
    text = sample["text"].replace('\n', ' ')
    length = len(text)
    if length < MIN_CHARS or length > MAX_CHARS:
        continue
    parts.append(text)
    total_chars += length
    if total_chars >= TARGET_CHARS:
        break

text_sequence = '\n'.join(parts)
print(f"Collected {len(text_sequence):,} characters from {len(parts):,} documents")
```

## BPE algorithm

I am using the [minBPE](https://github.com/karpathy/minbpe) repository to tokenize the sequence of text.

```python
import sys
sys.path.append('..')
```

Start by training the tokenizer on the text sequence that you saved in the previous notebook.

```python
from minbpe import RegexTokenizer
from variant_gpt.models import GPT2Config

config = GPT2Config()

tokenizer = RegexTokenizer()
tokenizer.train(text_sequence, vocab_size=config.vocab_size)
```

Visualize the vocabulary.

```python
vocab = tokenizer.vocab
vocab
```

Test the tokenizer.

```python
tokenizer.encode("Ciao mondo!")
```

```python
tokenizer.decode([13787, 826, 33])
```

Add special tokens to the vocabulary. These tokens are going to be used a lot in the fine-tuning step.

```python
max_vocab_id = list(tokenizer.vocab.keys())[-1]
tokenizer.register_special_tokens({
    "<|startoftext|>": max_vocab_id + 1,
    "<|separator|>":   max_vocab_id + 2,
    "<|endoftext|>":   max_vocab_id + 3,
    "<|unk|>":         max_vocab_id + 4,
    "<|padding|>":     max_vocab_id + 5,
})
```

Save the tokenizer

```python
tokenizer.save(file_prefix="../output/tokenizer/fineweb_tokenizer_2")
```
